```r
library(maftools)
```

```r
var.annovar <-"/home2/zhenglijing/work/SNP_bcf/dele_AD/public_data/all_sample_merge2.txt"
```

```r
var.annovar.maf = annovarToMaf(annovar = var.annovar, 
                               refBuild = 'hg19',
                               sep = ",",
                               tsbCol = 'Barcode',
                               removeNonMutated=TRUE,
                               MAFobj = TRUE)
 
oncoplot(maf = var.annovar.maf, draw_titv = TRUE)
```

```r
mutation_types <- c("Missense_Mutation")
mutation_colors <- c("Missense_Mutation" = "#9467bd", 'Frame_Shift_Del' = 'Red')
oncoplot(maf = var.annovar.maf,
         draw_titv = TRUE,
         colors = mutation_colors,
        gene_mar=7)
```
